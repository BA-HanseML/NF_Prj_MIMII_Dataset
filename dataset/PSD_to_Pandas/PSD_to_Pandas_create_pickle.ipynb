```python
# create frame
import time
t = time.time()
FileFindDict={'SNR': '6dB','machine': 'pump','ID': ['00','02','04','06']}
PSDname = 'full_spectrum_128'
pump = PSD_to_Pandas(PSD_window='hamming',
                     PSD_nperseg = 128,
                     PSD_nfft=512,
                     PSD_scaling='spectrum',
                     base_folder='../../', 
                     FileFindDict=FileFindDict, 
                     FileCountlimit=None, # !test with lesser this can take long time !
                     ChannelNr=0,
                     freq_band = [200,8000])

elapsed = time.time() - t
print(elapsed)
# save
FileName = FileFindDict['machine'] + '_' + \
           FileFindDict['SNR'] + '_' + \
           ''.join(FileFindDict['ID']) + '_' + PSDname + '.pkl'

pump.to_pickle(FileName)
```
